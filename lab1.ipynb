```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## Problema 1

```python
def colour_filter(image, colour):
    img = cv2.cvtColor(image,
                       cv2.COLOR_BGR2RGB)
    
    #Dimensiones
    
    alto = img.shape[0]
    ancho = img.shape[1]
    canales = img.shape[2]
    
    #colores
    
    colores = np.zeros_like(img)
    
    #pixeles
    
    for i in range(0,alto):
        for j in range(0,ancho):
            pix = img[i,j]
            blue = pix[0]
            green = pix[1]
            red = pix[1]
            
            #filtramos por color
            
            if colour == 1:
                colores[i,j] = (blue,0,0)
            elif colour == 2:
                colores[i,j] = (0,green,0)
            elif colour == 3:
                colores[i,j] = (0,0,red)
            elif colour == 10:
                colores[i,j] = (0,green,red)
            elif colour == 20:
                colores[i,j] = (blue,green,0)
            elif colour == 30:
                colores[i,j] = (blue,0,red)
    return colores
```

Par√°metro de color = 1

```python
img = cv2.imread("lotr.jpg")

color = 1

imagen_filtrada = colour_filter(img, color)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(imagen_filtrada)
plt.title('Filtro')
plt.axis('off')

plt.show()
```
