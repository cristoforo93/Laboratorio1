## Problema 1

```python
def colour_filter(image, colour):
    img = cv2.cvtColor(image,
                       cv2.COLOR_BGR2RGB)
    
    #Dimensiones
    
    alto = img.shape[0]
    ancho = img.shape[1]
    canales = img.shape[2]
    
    #colores
    
    colores = np.zeros_like(img)
    
    #pixeles
    
    for i in range(0,alto):
        for j in range(0,ancho):
            pix = img[i,j]
            blue = pix[0]
            green = pix[1]
            red = pix[1]
            
            #filtramos por color
            
            if colour == 1:
                colores[i,j] = (blue,0,0)
            elif colour == 2:
                colores[i,j] = (0,green,0)
            elif colour == 3:
                colores[i,j] = (0,0,red)
            elif colour == 10:
                colores[i,j] = (0,green,red)
            elif colour == 20:
                colores[i,j] = (blue,green,0)
            elif colour == 30:
                colores[i,j] = (blue,0,red)
    return colores
```

Par√°metro de color = 1

```python
img = cv2.imread("lotr.jpg")

color = 1

imagen_filtrada = colour_filter(img, color)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
plt.title('Original')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(imagen_filtrada)
plt.title('Filtro')
plt.axis('off')

plt.show()
```

## Problema 2

```python
def colour_image(imgr,imgg,imgb):
    
    #dimensiones
    
    if imgr.shape != imgg.shape or imgr.shape != imgb.shape:
        raise ValueError("Las im√°genes tienen distintos tama√±os.")
    
    #matrix 3D para la imagen a color
    
    img = np.zeros((imgr.shape[0],
                    imgg.shape[1],
                    3),
                   dtype=np.uint8)
    
    #canal de color
    
    img[:, :, 0] = imgb #canal azul
    img[:, :, 1] = imgg #canar verde
    img[:, :, 2] = imgr #canar rojo
    
    return img
```

Imagen 1

```python
image_red = cv2.imread("imagen1\imagen1_salida_gray_rojo.jpg",
                       cv2.IMREAD_GRAYSCALE)
image_green = cv2.imread("imagen1\imagen1_salida_gray_verde.jpg",
                         cv2.IMREAD_GRAYSCALE)
image_blue = cv2.imread("imagen1\imagen1_salida_gray_azul.jpg",
                        cv2.IMREAD_GRAYSCALE)

imagen_color = colour_image(image_red,
                            image_green,
                            image_blue)

plt.imshow(imagen_color)
plt.title("Imagen a Color")
plt.axis("off")
plt.show()
```

```python
def histogramas(imagen):
    
    hist_red = cv2.calcHist([imagen],
                            [0],
                            None,
                            [256],
                            [0,256])
    hist_green = cv2.calcHist([imagen],
                            [1],
                            None,
                            [256],
                            [0,256])
    hist_blue = cv2.calcHist([imagen],
                            [2],
                            None,
                            [256],
                            [0,256])
    
    total_pixeles = imagen.shape[0] * imagen.shape[1]
    media_r = np.sum(np.arange(256) * hist_red.flatten()) / total_pixeles
    media_g = np.sum(np.arange(256) * hist_green.flatten()) / total_pixeles
    media_b = np.sum(np.arange(256) * hist_blue.flatten()) / total_pixeles
    
    plt.figure(figsize=(12,8))
    
    plt.subplot(231)
    plt.plot(hist_red,color='red')
    plt.axvline(x=media_r,color='red',linestyle='dashed',ymin=0,ymax=0.9)
    plt.xlim([0,255])
    plt.title('Rojo')
    
    plt.subplot(232)
    plt.plot(hist_green,color='green')
    plt.axvline(x=media_g,color='green',linestyle='dashed',ymin=0,ymax=0.9)
    plt.xlim([0,255])
    plt.title('Verde')
    
    plt.subplot(233)
    plt.plot(hist_blue,color='blue')
    plt.axvline(x=media_b,color='blue',linestyle='dashed',ymin=0,ymax=0.9)
    plt.xlim([0,255])
    plt.title('Azul')
    
    plt.tight_layout()
    plt.show()
```

```python
imagen = cv2.imread('lotr.jpg')
histogramas(imagen)
```
